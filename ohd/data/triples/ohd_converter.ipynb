```python
import rdflib as rdf
from rdflib import Graph
graph  = Graph()
graph.parse(OUTPUT_PATH, format='xml')
```

```python
triples = [triple for triple in graph]
df = pd.DataFrame(triples, columns=['subject', 'predicate', 'object'])
df
```

```python
# map every values to string
df = df.applymap(lambda x: str(x))
df.to_csv(TRIPLE_PATH, index=False, header=False)
```

```python
subclass_df = df[df["predicate"] == 'http://www.w3.org/2000/01/rdf-schema#subClassOf']
subclass_df = subclass_df.drop(columns=["predicate"])
```

```python
def is_subclass_of(subclass, superclass):
    objects = subclass_df[subclass_df["subject"] == subclass]["object"].values
    for object in objects:
        if object == superclass:
            return True
        if is_subclass_of(object, superclass):
            return True
    return False
```

```python
from tqdm import tqdm
# generate negative samples
negative_samples = []
for subclass in tqdm(subclass_df["subject"].sample(len(subclass_df) // 3)):
    # sample 3 negative samples
    for _ in range(3):
        while True:
            superclass = subclass_df.sample(1)["object"].values[0]
            if not is_subclass_of(subclass, superclass):
                break
        negative_samples.append((subclass, "http://www.w3.org/2000/01/rdf-schema#subClassOf", superclass))
```

```python
negative_samples_df = pd.DataFrame(negative_samples, columns=['subject', 'predicate', 'object'])
# drop predicate column
negative_samples_df = negative_samples_df.drop(columns=["predicate"])
# split subclass_df into train and test
train_df = subclass_df.sample(frac=0.8)
test_df = subclass_df.drop(train_df.index)

# add columns of 1 (positive) to train_df and 0 (negative) to negative_samples_df
train_df["label"] = 1
negative_samples_df["label"] = 0 

# add negative samples to train_df
train_df = pd.concat([train_df, negative_samples_df])
train_df.to_csv("train.csv", index=False, header=False)
test_df.to_csv("test.csv", index=False, header=False)
test_df.to_csv("valid.csv", index=False, header=False) # tmp
```
