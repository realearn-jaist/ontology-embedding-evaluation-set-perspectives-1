```python
import random
import pickle
import numpy as np
import pandas as pd
from tqdm import tqdm
from pprint import pprint
from eval_util import Report 
report_man = Report()
```

```python
report = report_man.load()
models = list(report.keys())
```

```python
res_df = pd.DataFrame([report[model]["res"]["inclusive"] for model in models], index = models)
res_df["logSemMR"] = np.log2(res_df["SemMR"])
```

```python
res_df[["MR", "logSemMR", "MRR", "SemMRR", "hits1", "hits5", "hits10"]] \
    .sort_values("MR", ascending=False) \
    .style \
    .background_gradient(cmap='bwr') \
    .format("{:.4f}")
```

```python
from IPython.display import display

def display_report(model):
    print(model)
    
    print("res_class_entity_visualization..")
    display(report[model]["res"]["concept"]["res_class_entity_visualization"])
    
    print("res_cluster_visualization..")
    display(report[model]["res"]["concept"]["res_cluster_visualization"])
    
    print("concept2vec: task 1.1 Categorization metric..")
    display(pd.Series(report[model]["res"]["concept"]["res_task1"][0]))
    
    print("concept2vec: task 1.2 Coherence metric..")
    display(pd.Series(report[model]["res"]["concept"]["res_task1"][1]))
    
    print("positive and negative samples distance..")
    display(pd.Series(report[model]["res"]["distance"]))
    
    print("metrics..")
    display(pd.Series(report[model]["res"]["inclusive"]))
    
    print("-" * 50)
```

```python
for model in models:
    display_report(model)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

all_data = [report[model]["res"]["concept"]["res_task1"][0] for model in models]
all_data = pd.DataFrame(all_data, index=models)


plt.figure(figsize=(10, 20))
sns.heatmap(all_data.T, annot=True)
plt.title("concept2vec task 1.1: Categorization metric")
plt.show()
```

```python
all_new_data = []
for model in models:
    new_data = {}
    for key, values in report[model]["res"]["concept"]["res_task1"][1].items():
        new_data[key] = values[0]  
    all_new_data.append(new_data)

all_new_data = pd.DataFrame(all_new_data, index=models)

plt.figure(figsize=(10, 20))
sns.heatmap(all_new_data.T, annot=True)
plt.title("concept2vec: task 1.2: Coherence metric")
plt.show()
```
